```python
def plot_number_of_oscillations (peaks_tod, data_id, filename=None):
    
    #saving number of oscillations to an array (by number of peak maximums)
   # number_of_oscillations = [len(value['max_int']) for value in peaks_tod.values()]
    number_of_oscillations = [len(value['max_int']) if value['max_int'] else 0 for value in peaks_tod.values()]
    number_of_oscillations = np.array(number_of_oscillations)
    

    # Create the plot
    fig = plt.figure(figsize=[3, 2])
    binwidth = 1
    bin_edges = np.arange(0, 20 + binwidth, binwidth)  # Use bin edges

    # Plot histogram
    counts, bins, patches = plt.hist(number_of_oscillations, color='grey', edgecolor='black', linewidth=1.5, bins=bin_edges)

    # Calculate bin midpoints for xticks
    bin_midpoints = 0.5 * (bins[1:] + bins[:-1])

    # Add a vertical line at the mean
    plt.axvline(np.mean(number_of_oscillations), linewidth=2, color='royalblue', alpha=0.7, linestyle='dashed')

    # Adjust x and y axis ticks and limits
    plt.xticks(bin_midpoints, labels=np.arange(0, len(bin_midpoints)), fontsize=12)  # Center ticks on bars
    plt.xlim([0, 10])  # Adjust x-axis limit as needed
    plt.yticks(np.arange(0, 30, 3), fontsize=12)
    plt.ylim([0, 10])  # Adjust y-axis limit as needed

    # Titles and labels
    plt.title(data_id, fontsize=12)
    plt.xlabel('Number of Oscillations', fontsize=12)
    plt.ylabel('Number of Cells', fontsize=12)

    # Save the figure if a filename is provided
    if filename:
        plt.savefig(filename, dpi=300, bbox_inches='tight', format='png')
        save_path = os.path.abspath(filename)
        print(f"Figure saved to: {save_path}")
    
    plt.show()

    return
```

### Period per cycle

```python
def plot_period_per_cycle(peaks_tod, data_id=None,filename=None,for_cycles=None,reverse_cycles=False):
    '''
    this function plots the period between maximum peaks. in default mode, it plots the period per cycle 
    when data_id='xx' is added, title becomes xx
    when filename='xx.png' is added, it exports the plot 
    when for_cycles=n is added, it plots only the cells that show n+1 oscillations 
    when reverse_cycles=True, it aligns last peaks 
    '''

    #get lists of all max peak timings
    peaks = [value['peaks']['max_time'] for value in peaks_tod.values()]

    # Function to compute differences between consecutive elements
    def compute_period(mylist):
        return [j - i for i, j in zip(mylist[:-1], mylist[1:])]

    # Applying the function to each list in peaks
    
    period_per_cycle = [compute_period(sublist) for sublist in peaks]
    if for_cycles is not None:
         period_per_cycle = [compute_period(sublist) for sublist in peaks if len(sublist)==for_cycles]
    
    #add an option to reverse the lists to plot starting from the last cycle backwards
    if reverse_cycles==True:
        def compute_period(mylist):
        # Example implementation: compute differences between consecutive elements
            return [i - j for i, j in zip(mylist[:-1], mylist[1:])]
        period_per_cycle = [compute_period(sublist[::-1]) for sublist in peaks]
        

    grouped_values = {}

    # Group values by their position in the sublists
    for sublist in period_per_cycle:
        for idx, value in enumerate(sublist):
            if idx + 1 not in grouped_values:
                grouped_values[idx + 1] = []
            grouped_values[idx + 1].append(value)

    # Plot the grouped values
    plt.figure(figsize=(3, 2))

    for idx, group in grouped_values.items():
        # Add jitter to the x-values
        x_values = np.array([idx] * len(group)) + np.random.uniform(-0.1, 0.1, len(group))
        plt.scatter(x_values, group, edgecolors='royalblue', alpha=0.5, facecolors='none', label=f'Index {idx}' if idx == 1 else"")

    # Calculate means and standard deviations
    mean_values = []
    std_values = []
    x_values = []
    for idx in sorted(grouped_values.keys()):
        group = grouped_values[idx]
        mean_values.append(np.mean(group))
        std_values.append(np.std(group))
        x_values.append(idx)

    
    plt.errorbar(x_values, mean_values, yerr=std_values, fmt='-', linewidth=1, 
                 color='black', ecolor='black', elinewidth=1, capsize=4)

    plt.xlabel('cycle', fontsize=12)
    plt.xticks(np.arange(1,9,1),fontsize=12)
    plt.ylabel('period (min)', fontsize=12)
    plt.ylim(40,180)
    plt.yticks(np.arange(40,200,20),fontsize=12)
    plt.title(data_id, fontsize=12)
    save_path=None
    if filename:
        plt.savefig(filename, dpi=300, bbox_inches='tight', format='png')
        # Print the save path
        save_path = os.path.abspath(filename)
    print(f"Figure saved to: {save_path}")
    
    plt.show()

    return
```

### Average period

```python
def average_period(peaks_tod, data_id=None):

    #get lists of all max peak timings
    peaks = [value['peaks']['max_time'] for value in peaks_tod.values()]

    # Function to compute differences between consecutive elements
    def compute_period(mylist):
        return [j - i for i, j in zip(mylist[:-1], mylist[1:])]

    period_per_cycle = [compute_period(sublist) for sublist in peaks]
    # Calculate the average for each sublist in period_per_cycle
    averages = [np.mean(sublist) for sublist in period_per_cycle]
    return averages
```

```python
# Plot the averages in a violin plot
plt.figure(figsize=(3, 2))
sns.violinplot(data=[average_period(dmso), average_period(lpa)], inner="quart", color="royalblue") 
    # inner="point" / inner="quart" 

plt.ylabel('average period (min)')
plt.xticks(ticks=[0, 1], labels=["DMSO", "LPA (100 nM)"])

# Save the plot as a PNG file
plt.savefig("lpa-average-period.png", dpi=300, bbox_inches='tight', format='png')
plt.show()
```

```python
# Plot the averages in a violin plot
plt.figure(figsize=(3, 2))
sns.violinplot(data=[average_period(dmso), average_period(pnbb25), average_period(pnbb50)], inner="quart", color="royalblue") 
    # inner="point" / inner="quart" 

plt.ylabel('average period (min)')
plt.xticks(ticks=[0, 1, 2], labels=["DMSO", "pnbb \n(25 uM)", "pnbb \n(50 uM)"])

# Adjust the layout to make sure everything fits
plt.tight_layout()
# Save the plot as a PNG file
plt.savefig("pnbb-average-period.png", dpi=300, bbox_inches='tight', format='png')
plt.show()
```

Statistics

```python
from scipy.stats import shapiro, f_oneway, kruskal
import numpy as np

# Perform Shapiro-Wilk test for each group
shapiro_dmso = shapiro(dmso)
shapiro_pnbb_25uM = shapiro(pnbb25)
shapiro_pnbb_50uM = shapiro(pnbb50)

print("Shapiro-Wilk Test Results:")
print("DMSO:", shapiro_dmso)
print("pnbb (25 uM):", shapiro_pnbb_25uM)
print("pnbb (50 uM):", shapiro_pnbb_50uM)

# Check normality and choose the appropriate test
if (shapiro_dmso.pvalue > 0.05 and
    shapiro_pnbb_25uM.pvalue > 0.05 and
    shapiro_pnbb_50uM.pvalue > 0.05):
    # Perform One-Way ANOVA if data is normally distributed
    anova_result = f_oneway(dmso, pnbb25, pnbb50)
    print("\nOne-Way ANOVA result:", anova_result)
else:
    # Perform Kruskal-Wallis Test if data is not normally distributed
    kruskal_result = kruskal(dmso, pnbb25, pnbb50)
    print("\nKruskal-Wallis test result:", kruskal_result)
```
